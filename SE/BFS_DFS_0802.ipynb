```python
# 재귀함수 - 스택 자료구조를 활용할 때 많이 사용함. 대표적인 예 - DFS

# 2가지 방식으로 구현한 팩토리얼 예제

def factorial_iterative(n):
  result = 1
  for i in range(1,n+1):
    result *= i
  return result

def factorial_recursive(n):
  if n<=1:
    return 1
  return n*factorial_recursive(n-1)

print(factorial_iterative(5))
print(factorial_recursive(5))
```

### DFS 예제
- 숫자가 낮은 노드부터 시작하여 방문한다.
- 방문한 노드는 스택에 쌓는다.
- 재귀함수를 쓰는게 특징. 끝까지 탐색한 후 빠져나올 조건을 잘 줘야한다.

```python
visited = [False]*9 #방문 처리 할 스택

def dfs(graph,v,visited):
  visited[v] = True # 방문 처리
  print(v, end=' ')
  
  for i in graph[v]:
    if not visited[i]:
      dfs(graph,i,visited)

for i in graph[1]([2,3,8]):
  if not visited[i]:
    dfs(graph,i,visited)

# 인접 리스트
graph =[
        [],
        [2,3,8],
        [1,7],
        [1,4,5],
        [3,5],
        [3,4],
        [7],
        [2,6,8],
        [1,7]
]

# dfs(graph,1,visited)
graph
```

```python
graph = [[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7]]
visited = [False]*9

# 똑같은 방식의 탐색 반복 -> 재귀 함수 사용하는거 잊지 말기!
def DFS(graph,v,visited):
  visited[v] = True
  print(v, end=' ')

  for i in graph[v]:
    if not visited[i]:
      DFS(graph,i,visited)

DFS(graph,1,visited)
```

```python
from collections import deque

graph = [[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7]]
visited = [False]*9

def BFS(graph,start,visited):
  queue = deque([start])
  visited[start] = True
  while queue:
    v = queue.popleft() #처음에 넣었던거 빼고
    print(v, end=' ') #방문했으니까 print
    for i in graph[v]: #v의 인접 노드들 확인
      if not visited[i]: #방문하지 않았다면
        queue.append(i) #방문하지 않은 인접 노드들 모두 queue에 추가하기
        visited[i] = True 

BFS(graph,1,visited)
```
